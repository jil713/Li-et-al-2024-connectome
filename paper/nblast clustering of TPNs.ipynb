```python
import navis
```

```python
#Import swc files
skeletons = navis.read_swc (r"D:\Gordon Lab\codex SEZ TPN project\Paper preparation\paper_final raw data\connectivity\ACh interneuron vs 2nd projection\swc for nblast analysis\v783\all", include_subdirs=True)
```

```python
#Change skeleton names into IDs
import numpy as np
labels=skeletons.name
labelslst=labels.tolist()
skeletons.set_neuron_attributes(labelslst, 'id', register=True, na='raise')
```

```python
#Convert skeleton unit into um
nl_um = skeletons / 1000
```

```python
#Make dotprops
dps = navis.make_dotprops(nl_um, k=5, resample=False)
nbl = navis.nblast(dps[:3], dps[3:], progress=False)
```

```python
#Generate all-by-all comparisons
aba = navis.nblast_allbyall(dps, progress=False)
```

```python
#Average forward and reverse scores
aba_mean = (aba + aba.T) / 2
```

```python
#Compute distance for hierarchical clustering
aba_dist = 1 - aba_mean
```

```python
#Perform hierarchical clustering and generate dendrogram
from scipy.spatial.distance import squareform
from scipy.cluster.hierarchy import linkage, dendrogram, set_link_color_palette

import matplotlib.pyplot as plt
import matplotlib.colors as mcl
import seaborn as sns

set_link_color_palette([mcl.to_hex(c) for c in sns.color_palette('muted', 10)])
#Change to vector-form
aba_vec = squareform(aba_dist, checks=False)
Z = linkage(aba_vec, method='ward')
plt.figure(figsize=(30, 5))
dendrogram_result=dendrogram(Z, labels=aba_mean.columns, leaf_rotation=90, leaf_font_size=10)
plt.tight_layout()
#plt.savefig("TPN morphology clustering.pdf", format="pdf", dpi=300)
```
